```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np


categories = {
    "anxiety": {
        "keywords": ["anxious", "worried", "panic", "fear", "nervous"],
        "anchors": [
            "I feel anxious and worried",
            "I am scared about what might happen",
            "I feel nervous and panicked"
        ]
    },
    "depression": {
        "keywords": ["hopeless", "empty", "sad", "worthless", "depressed"],
        "anchors": [
            "I feel hopeless and empty",
            "I feel sad and worthless",
            "I have no motivation to continue"
        ]
    },
    "stress": {
        "keywords": ["stress", "pressure", "overwhelmed", "tension", "burden"],
        "anchors": [
            "I feel stressed and overwhelmed",
            "I am under too much pressure",
            "Everything feels like a burden"
        ]
    },
    "burnout": {
        "keywords": ["exhausted", "drained", "tired", "fatigue", "burned out"],
        "anchors": [
            "I feel mentally exhausted",
            "I am emotionally drained",
            "I feel burned out from everything"
        ]
    }
}


model = SentenceTransformer("all-MiniLM-L6-v2")


def lexical_score(text, keywords):
    text = text.lower()
    score = 0

    for word in keywords:
        if word.lower() in text:
            score += 1

    return score


def build_category_centroids(categories):
    centroids = {}

    for category, data in categories.items():
        anchor_embeddings = model.encode(data["anchors"])
        centroids[category] = np.mean(anchor_embeddings, axis=0)

    return centroids


category_centroids = build_category_centroids(categories)


def weak_label(text, alpha=0.4, beta=0.6):
    scores = {}

    text_embedding = model.encode([text])[0]

    for category, data in categories.items():
        lex_score = lexical_score(text, data["keywords"])

        sem_score = cosine_similarity(
            [text_embedding],
            [category_centroids[category]]
        )[0][0]

        final_score = alpha * lex_score + beta * sem_score

        scores[category] = {
            "lexical_score": lex_score,
            "semantic_score": sem_score,
            "final_score": final_score
        }

    final_label = max(scores, key=lambda c: scores[c]["final_score"])

    return final_label, scores


text = "I feel completely drained and exhausted after work every day."

label, scores = weak_label(text)

print("Text:", text)
print("Predicted Label:", label)

for category, score in scores.items():
    print(category, score)
```

```python
import json
from datasets import Dataset, DatasetDict

def build_dpo_dataset(path):
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)

    rows = []

    for item in data:
        instruction = item.get("instruction", "").strip()
        category = item.get("category", "general emotional support").strip()

        chosen = item.get("chosen") or item.get("good_response")
        rejected = item.get("rejected") or item.get("bad_response")

        if not instruction or not chosen or not rejected:
            continue

        prompt = (
            "<|begin_of_text|><|start_header_id|>system<|end_header_id|>\n"
            "You are an emotionally supportive mental health chatbot. "
            "Respond with empathy, validation, emotional clarity, and safe supportive guidance. "
            "Do not provide medical diagnosis.\n"
            "<|eot_id|><|start_header_id|>user<|end_header_id|>\n"
            f"Category: {category}\n"
            f"{instruction}\n"
            "<|eot_id|><|start_header_id|>assistant<|end_header_id|>\n"
        )

        rows.append({
            "prompt": prompt,
            "chosen": chosen.strip(),
            "rejected": rejected.strip()
        })

    return Dataset.from_list(rows)

# Build dataset
full_dataset = build_dpo_dataset(DATA_PATH)

# First split: Train (80%) and Temp (20%)
split_1 = full_dataset.train_test_split(
    test_size=0.2,
    seed=42
)

train_dataset = split_1["train"]
temp_dataset = split_1["test"]

# Second split: Validation (10%) and Test (10%)
split_2 = temp_dataset.train_test_split(
    test_size=0.5,
    seed=42
)

validation_dataset = split_2["train"]
test_dataset = split_2["test"]

# Final dataset dictionary
dataset = DatasetDict({
    "train": train_dataset,
    "validation": validation_dataset,
    "test": test_dataset
})

print(dataset)

print("\nTrain size:", len(dataset["train"]))
print("Validation size:", len(dataset["validation"]))
print("Test size:", len(dataset["test"]))

print("\nSample Train Record:")
print(dataset["train"][0])
```

```python
from trl import DPOTrainer, DPOConfig
import torch

training_args = DPOConfig(
    output_dir="/content/llama3_2_3b_dpo_unsloth",

    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,

    gradient_accumulation_steps=16,

    num_train_epochs=1,
    learning_rate=3e-6,
    warmup_ratio=0.05,
    beta=0.1,

    logging_steps=10,

    # Validation after every 10 training steps
    eval_strategy="steps",
    eval_steps=10,

    # Save checkpoint after every 10 training steps
    save_strategy="steps",
    save_steps=10,

    fp16=not torch.cuda.is_bf16_supported(),
    bf16=torch.cuda.is_bf16_supported(),

    optim="adamw_8bit",

    report_to="none",
    remove_unused_columns=False,
)

trainer = DPOTrainer(
    model=model,
    ref_model=None,
    args=training_args,

    train_dataset=dataset["train"],
    eval_dataset=dataset["validation"],

    processing_class=tokenizer,

    max_prompt_length=128,
    max_length=256,
)

trainer.train()
```

```python
from unsloth import FastLanguageModel
import torch

MODEL_PATH = "/content/llama3_2_3b_dpo_adapter"
max_seq_length = 512

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = MODEL_PATH,
    max_seq_length = max_seq_length,
    dtype = None,
    load_in_4bit = True,
)

FastLanguageModel.for_inference(model)
```

```python
def generate_response(instruction, category):
    prompt = (
        "<|begin_of_text|><|start_header_id|>system<|end_header_id|>\n"
        "You are an emotionally supportive mental health chatbot. "
        "Respond with empathy, validation, emotional clarity, and safe supportive guidance. "
        "Do not provide medical diagnosis.\n"
        "<|eot_id|><|start_header_id|>user<|end_header_id|>\n"
        f"Category: {category}\n"
        f"{instruction}\n"
        "<|eot_id|><|start_header_id|>assistant<|end_header_id|>\n"
    )

    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

    outputs = model.generate(
        **inputs,
        max_new_tokens=120,
        temperature=0.7,
        top_p=0.9,
        do_sample=True,
        eos_token_id=tokenizer.eos_token_id,
    )

    decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)

    if "assistant" in decoded.lower():
        return decoded.split("assistant")[-1].strip()

    return decoded[len(prompt):].strip()
```

```python
test_data = dataset["test"]

predictions = []
references = []
categories = []
instructions = []

for item in test_data:
    instruction = item["prompt"].split("<|start_header_id|>user<|end_header_id|>")[-1]
    instruction = instruction.split("<|eot_id|>")[0].strip()

    category = "general emotional support"
    user_text = instruction

    if instruction.startswith("Category:"):
        lines = instruction.split("\n")
        category = lines[0].replace("Category:", "").strip()
        user_text = "\n".join(lines[1:]).strip()

    pred = generate_response(user_text, category)

    predictions.append(pred)
    references.append(item["chosen"])
    categories.append(category)
    instructions.append(user_text)

print("Generated:", len(predictions))
print(predictions[0])
```

```python
import evaluate
import numpy as np

bleu = evaluate.load("sacrebleu")
rouge = evaluate.load("rouge")
bertscore = evaluate.load("bertscore")

bleu_result = bleu.compute(
    predictions=predictions,
    references=[[ref] for ref in references]
)

rouge_result = rouge.compute(
    predictions=predictions,
    references=references
)

bert_result = bertscore.compute(
    predictions=predictions,
    references=references,
    lang="en"
)

avg_bert_f1 = np.mean(bert_result["f1"])
avg_response_length = np.mean([len(p.split()) for p in predictions])

print("BLEU:", bleu_result["score"])
print("ROUGE-1:", rouge_result["rouge1"])
print("ROUGE-2:", rouge_result["rouge2"])
print("ROUGE-L:", rouge_result["rougeL"])
print("BERTScore F1:", avg_bert_f1)
print("Average response length:", avg_response_length)
```

```python
import pandas as pd
import numpy as np

# Create dataframe
results_df = pd.DataFrame({
    "instruction": instructions,
    "category": categories,
    "reference_good_response": references,
    "generated_response": predictions,
})

# Save predictions
results_df.to_csv("/content/dpo_test_predictions.csv", index=False)

# Save metrics summary
summary_df = pd.DataFrame([{
    "BLEU": bleu_result["score"],
    "ROUGE-1": rouge_result["rouge1"],
    "ROUGE-2": rouge_result["rouge2"],
    "ROUGE-L": rouge_result["rougeL"],
    "BERTScore_F1": float(avg_bert_f1),
    "Average_Response_Length": float(avg_response_length),
}])

summary_df.to_csv("/content/dpo_metric_summary.csv", index=False)

print("Files saved successfully.")
```

```python
from unsloth import FastLanguageModel
import torch

BASE_MODEL_NAME = "unsloth/Llama-3.2-3B-Instruct-bnb-4bit"

base_model, base_tokenizer = FastLanguageModel.from_pretrained(
    model_name=BASE_MODEL_NAME,
    max_seq_length=512,
    dtype=None,
    load_in_4bit=True,
)

FastLanguageModel.for_inference(base_model)
```

```python
def generate_base_response(instruction, category):
    prompt = (
        "<|begin_of_text|><|start_header_id|>system<|end_header_id|>\n"
        "You are an emotionally supportive mental health chatbot. "
        "Respond with empathy, validation, emotional clarity, and safe supportive guidance. "
        "Do not provide medical diagnosis.\n"
        "<|eot_id|><|start_header_id|>user<|end_header_id|>\n"
        f"Category: {category}\n"
        f"{instruction}\n"
        "<|eot_id|><|start_header_id|>assistant<|end_header_id|>\n"
    )

    inputs = base_tokenizer(prompt, return_tensors="pt").to("cuda")

    with torch.no_grad():
        outputs = base_model.generate(
    **inputs,
    max_new_tokens=60,
    temperature=0.7,
    top_p=0.9,
    do_sample=True,
    eos_token_id=base_tokenizer.eos_token_id,
    pad_token_id=base_tokenizer.eos_token_id,
        )

    generated_ids = outputs[0][inputs["input_ids"].shape[-1]:]
    response = base_tokenizer.decode(generated_ids, skip_special_tokens=True)

    return response.strip()
```

```python
base_predictions = []
base_references = []
base_categories = []
base_instructions = []

# Use only 30 samples for faster baseline evaluation
sample_size = min(30, len(dataset["test"]))
test_subset = dataset["test"].select(range(sample_size))

for idx, item in enumerate(test_subset, start=1):
    prompt_text = item["prompt"]

    user_part = prompt_text.split("<|start_header_id|>user<|end_header_id|>")[-1]
    user_part = user_part.split("<|eot_id|>")[0].strip()

    category = "general emotional support"
    instruction = user_part

    if user_part.startswith("Category:"):
        lines = user_part.split("\n")
        category = lines[0].replace("Category:", "").strip()
        instruction = "\n".join(lines[1:]).strip()

    pred = generate_base_response(instruction, category)

    base_predictions.append(pred)
    base_references.append(item["chosen"])
    base_categories.append(category)
    base_instructions.append(instruction)

    print(f"Generated {idx}/{sample_size}")

print("Generated base responses:", len(base_predictions))
print(base_predictions[0])
```

```python
from unsloth import FastLanguageModel
from transformers import TextStreamer
import torch

model_name = "unsloth/Llama-3.2-3B-Instruct"
adapter_path = "/content/llama3_2_3b_dpo_adapter/content/llama3_2_3b_dpo_adapter"

max_seq_length = 2048

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = model_name,
    max_seq_length = max_seq_length,
    dtype = None,
    load_in_4bit = True,
)

model.load_adapter(adapter_path)
FastLanguageModel.for_inference(model)

def generate_response(user_text, category):
    messages = [
        {
            "role": "system",
            "content": (
                "You are an empathetic mental health support chatbot. "
                "Respond supportively and safely. Do not give medication, diagnosis, "
                "or clinical treatment advice."
            )
        },
        {
            "role": "user",
            "content": f"Category: {category}\nUser: {user_text}"
        }
    ]

    inputs = tokenizer.apply_chat_template(
        messages,
        tokenize=True,
        add_generation_prompt=True,
        return_tensors="pt"
    ).to("cuda")

    outputs = model.generate(
        input_ids=inputs,
        max_new_tokens=200,
        temperature=0.7,
        top_p=0.9,
        do_sample=True,
        eos_token_id=tokenizer.eos_token_id
    )

    response = tokenizer.decode(outputs[0][inputs.shape[-1]:], skip_special_tokens=True)
    return response.strip()


user_text = "I feel exhausted and I don't know how to handle everything."
category = "stress"

print(generate_response(user_text, category))
```
